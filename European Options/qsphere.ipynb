```python
from qiskit.circuit import QuantumCircuit, QuantumRegister
from qiskit.quantum_info import Statevector
import matplotlib
%matplotlib inline
```

```python
qreg = QuantumRegister(2, 'q')
```

```python
circuit = QuantumCircuit(qreg)
```

```python
circuit.draw('mpl')
```

```python
state = Statevector(circuit)
state.draw('qsphere')
```

```python
circuit.h(0)
```

```python
circuit.draw('mpl')
```

```python
state = Statevector(circuit)
state.draw('qsphere')
```

```python
circuit.cx(0, 1)
```

```python
circuit.draw('mpl')
```

```python
state = Statevector(circuit)
state.draw('qsphere')
```
