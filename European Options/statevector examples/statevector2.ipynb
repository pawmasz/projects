```python
from qiskit.circuit import QuantumCircuit, QuantumRegister
from qiskit.quantum_info import Statevector
import matplotlib
%matplotlib inline
```

```python
qreg = QuantumRegister(1, 'q')
```

```python
circuit = QuantumCircuit(qreg)
circuit.h(0)
```

```python
state = Statevector(circuit)
state.draw('bloch')
```

```python
state.draw('text')
```

```python
state.draw('latex')
```

```python
circuit.t(0)
```

```python
circuit.draw('mpl')
```

```python
state = Statevector(circuit)
state.draw('bloch')
```

```python
state.draw('text')
```

```python
state.draw('latex')
```
