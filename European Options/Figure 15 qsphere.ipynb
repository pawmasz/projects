```python
from qiskit.circuit import QuantumCircuit, QuantumRegister
from qiskit.quantum_info import Statevector
import matplotlib
%matplotlib inline
```

```python
qreg = QuantumRegister(3, 'q')
```

```python
circuit = QuantumCircuit(qreg)
```

```python
circuit.draw('mpl')
```

```python
state = Statevector(circuit)
state.draw('qsphere')
```

```python
circuit.ry(1.43, 0)
circuit.ry(2.90, 1)
```

```python
circuit.draw('mpl')
```

```python
state = Statevector(circuit)
state.draw('qsphere')
```

```python
circuit.cx(0, 1)
```

```python
circuit.draw('mpl')
```

```python
state = Statevector(circuit)
state.draw('qsphere')
```

```python
circuit.ry(1.10, 0)
circuit.ry(-0.93, 1)
```

```python
circuit.draw('mpl')
```

```python
state = Statevector(circuit)
state.draw('qsphere')
```
